### Data Cleaning

```python
df.columns=df.columns.str.lower().str.replace(" ","_")
df.columns
```

```python
df.drop_duplicates(inplace=True,ignore_index=True)
```

```python
df.duplicated().sum()
```

```python
df.drop(['url','course_id'],axis=1,inplace=True)
```

```python
for i in df.select_dtypes(exclude="number"):
    print(i)
    print(df[i].nunique())
    print(df[i].unique())
    print("="*100)
```

```python
df['is_paid']=df['is_paid'].astype(str)
```

```python
def update_paid(x):
    if x=="True":
        return x.replace("True","paid")
    else :
        return x.replace("False","unpaid")
    
    
df['is_paid']=df['is_paid'].apply(update_paid) 
df['is_paid'] 
```

```python
print(df['is_paid'].unique())
print("\n",df['is_paid'].value_counts())
```

```python
df['published_timestamp']=pd.to_datetime(df['published_timestamp'])
```

```python
df['year'] = df['published_timestamp'].dt.year
df['month'] = df['published_timestamp'].dt.month
df['day'] = df['published_timestamp'].dt.day
```

```python
df
```

```python
for col in df.select_dtypes(include="number").columns:
    plt.figure(figsize=(6, 1.5))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot for {col}')
    plt.tight_layout()
    plt.show()
```

```python
from datasist.structdata import detect_outliers

feautre=['price','num_subscribers','num_reviews','num_lectures','content_duration']
outliers_cols = detect_outliers(df,n=0,features=feautre)
print(len(outliers_cols))
print(len(outliers_cols)/df.shape[0]*100)
```

```python
df.drop(outliers_cols,inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

### Data Analysis

```python
df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x='level',data=df)
plt.xlabel('Levels',fontsize=12)
plt.ylabel('Courses Count',fontsize=12)
plt.title('Levels vs Courses Count')
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.countplot(x='subject',data=df)
plt.xlabel('Subjects',fontsize=12)
plt.ylabel('Courses Count',fontsize=12)
plt.title('Subjects vs Courses Count')
plt.show()
```

```python
plt.figure(figsize=(8,7))
df['is_paid'].value_counts().plot(kind='bar')
plt.xlabel('Is_Paid',fontsize=12)
plt.ylabel('Courses Count',fontsize=12)
plt.title('Paid vs Free Courses')
plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.pie(x=df['subject'].value_counts().values,
        labels=df['subject'].value_counts().index,
        autopct="%.2f%%",
        explode=[.2,.1,0,0],
        shadow=True
       )
plt.show()
```

```python

counts = df.groupby('year')['is_paid'].value_counts().unstack().fillna(0)
print(counts)
```

```python
import matplotlib.pyplot as plt

counts.plot(kind='bar', stacked=True, color=['blue', 'red'], figsize=(10, 6))
plt.title('paid courses vs Year')
plt.xlabel('Year')
plt.ylabel('Counts courses')
plt.legend(['Paid', 'Unpaid'])
plt.show()
```

```python
counts = df.groupby('month')['is_paid'].value_counts().unstack()
print(counts)
```

```python
import matplotlib.pyplot as plt

counts.plot(kind='bar', stacked=True, color=['blue', 'red'], figsize=(10, 6))
plt.title('paid courses vs Year')
plt.xlabel('Year')
plt.ylabel('Counts courses')
plt.legend(['Paid', 'Unpaid'])
plt.show()
```

```python
sns.lmplot(x='price', y='num_subscribers', data=df, height=6, aspect=1.5)
plt.title('Price vs Num of Subscribers')
plt.xlabel(' Price')
plt.ylabel('Subscribers')
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.scatterplot(x='price',y='num_subscribers',data=df)
plt.xlabel('Price',fontsize=11)
plt.ylabel('Number of Subscribers',fontsize=11)
plt.title('Price vs Num of Subscribers')
plt.show()
```
