## Streamlit

```python
%%writefile fifa.py

import pandas as pd
import numpy as np
import plotly.express as px
import streamlit as st


st.set_page_config(layout="wide",page_title="Fifa Dashboard")
#read data 
df=pd.read_csv("clean fifa data.csv",index_col=0)

#create 3 pages
pages=st.sidebar.radio("Pages \n",['Overview','Player Analysis','Data Visualization'])

if pages=='Overview':
    st.markdown("<h1 style='text-align: center; color: white;'>Fifa Analysis</h1>", unsafe_allow_html=True)
    st.write("### Data sample")
    st.dataframe(df.head(10))
    st.metric("Shape  of Data", df.shape[0])
    st.write("### Is there any strong correlation between numerical columns")
    st.plotly_chart(px.imshow(df.select_dtypes(exclude="object").corr(),width=1000 , height=600))
    
elif pages=='Player Analysis':    
    st.markdown("<h1 style='text-align: center; color: white;'>Player Analysis</h1>", unsafe_allow_html=True)
    st.subheader("What are the top 10 most common nationalities among the players in dataset?")
    top_nationalities = df['nationality'].value_counts().head(10).reset_index()
    st.plotly_chart(px.bar(top_nationalities, x= 'nationality', y= 'count'))
    
    st.subheader("What is the distribution of players based on  positions  in  dataset?")
    q6=df["position"].value_counts().reset_index()
    st.plotly_chart(px.bar(q6,x="position",y="count"))
    
    st.subheader("What is the most common number of skill_moves among players in dataset?")
    q6=df["skill_moves"].value_counts().reset_index()
    st.plotly_chart(px.bar(q6,x="skill_moves",y="count"))
    
    st.subheader("What is the distribution of players based on preferred foot")
    q4=df['preferred_foot'].value_counts().reset_index()
    st.plotly_chart(px.pie(q4,names="preferred_foot",values="count"))
    
    st.write('### What distribution of age ?')
    st.plotly_chart(px.histogram(df, x= 'age'))
    
    st.write('### What distribution of wage ?')
    st.plotly_chart(px.histogram(df, x= 'wage'))
    
    st.write('### What distribution of value ?')
    st.plotly_chart(px.histogram(df, x= 'value'))
    
    st.subheader("What is the highest  players based on overall in dataset?")
    q90 = df.groupby("name")["overall"].max().sort_values(ascending=False).head(15).reset_index()
    st.plotly_chart(px.bar(q90,x='name',y='overall'))
    
    st.subheader("What is the lowest players based on overall in dataset?")
    q90 = df.groupby("name")["overall"].min().sort_values(ascending=False).head(15).reset_index()
    st.plotly_chart(px.bar(q90,x='name',y='overall'))
    
    st.subheader("What is the relationship  between age and value in dataset?")
    st.plotly_chart(px.scatter(df,x='age',y='value'))
    
    st.subheader("What is the relationship  between age and wage in dataset?")
    st.plotly_chart(px.scatter(df,x='age',y='wage'))
    
    st.subheader("What is the relationship  between age and skill moves in dataset?")
    st.plotly_chart(px.scatter(df,x='age',y='skill_moves'))
    
    
elif pages=='Data Visualization':
    st.markdown("<h1 style='text-align: center; color: white;'>Data Visualizations</h1>", unsafe_allow_html=True)
    s1=st.selectbox("Choose first columns",df.columns)
    s2=st.selectbox('Choose second column', df.columns)
    
    chart_type=st.selectbox('Choose Visualization chart', ['Histogram','Box Plot', 'Violin Plot', 'Strip Plot', 'Scatter'])
    
    if chart_type=='Histogram':
        st.plotly_chart(px.histogram(df, x = s1)) 
        st.plotly_chart(px.histogram(df, x = s2))
    elif chart_type == 'Box Plot':
        st.plotly_chart(px.box(df, x = s1, y= s2))
    elif chart_type == 'Violin Plot':
        st.plotly_chart(px.violin(df, x = s1, y= s2))
    elif chart_type == 'Strip Plot':
        st.plotly_chart(px.strip(df, x = s1, y= s2))
    else : 
        st.plotly_chart(px.scatter(df, x = s1, y= s2))
        
    
    

    

    
```
